# *Importação dos dados*



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja4.head()
```

# *Funções importantes*

```python
# Soma todos os valores de acordo com o dataframe e a coluna específica
def somarValores(dataframe, coluna):
  df = pd.DataFrame(dataframe)
  total = df[coluna].sum()
  return total

# Retorna duas listas
def listarItens(dataframe, coluna):
  df = pd.DataFrame(dataframe)
  itens = df[coluna].value_counts()
  indices_itens = itens.index.to_list()
  valores_itens = itens.to_list()
  listas = [indices_itens, valores_itens]
  return listas

# Exibe uma lista de índices e valores na tela
def imprimirItens(listas):
  indices_itens = listas[0]
  valores_itens = listas[1]
  for indices_itens,valores_itens in zip(indices_itens,valores_itens):
    print(f'{indices_itens:<25}\t{valores_itens:>10}')

# Calcula a média de itens em uma coluna
def calcularMedia(dataframe, coluna):
    df = pd.DataFrame(dataframe)
    total_item = somarValores(dataframe,coluna)
    quantidade_item = len(df[coluna])
    media = total_item/quantidade_item
    return round(media,2)
```

# *Listas importantes*

```python
# lista de lojas
lojas = [loja,loja2,loja3,loja4]
lojas_descricao = ['loja 1','loja 2','loja 3','loja 4']

# lista de faturamentos
faturamentos = []

# lista de média de avaliações
media_avaliacoes =[]

# lista de custo médio do frete
cm_frete = []

# lista das vendas por categoria
categoria_vendas = []

# lista das vendas por produto
produto_vendas = []
```


#1. Análise do faturamento

```python
for i in range(len(lojas)):
  faturamentos.append(somarValores(lojas[i],'Preço'))
  print(f'Faturamento da loja {i+1}: {faturamentos[i]}')
```

```python
# Calculando a média dos faturamentos
media = round(sum(faturamentos)/len(faturamentos),2)
media
```

# 2. Vendas por Categoria


```python
for i in range(len(lojas)):
  vendas = listarItens(lojas[i],'Categoria do Produto')
  categoria_vendas.append(vendas)

  print(f'Vendas por categoria da loja {i+1}')
  print(f'{"Produto":<25}\t{"Quantidade vendida":>10}')
  imprimirItens(vendas)
  print('\n')
```

# 3. Média de Avaliação das Lojas

```python
for i in range(len(lojas)):
  media_avaliacoes.append(calcularMedia(lojas[i],'Avaliação da compra'))
  print(f'Média de avaliações da loja {i+1}: {media_avaliacoes[i]}')
```

# 4. Produtos mais vendidos e produtos menos vendidos

```python
# Produtos vendidos por loja
for i in range(len(lojas)):
  vendas = listarItens(lojas[i],'Produto')
  produto_vendas.append(vendas)

  print(f'Vendas por produto da loja {i+1}')
  print(f'{"Produto":<25}\t{"Quantidade vendida":>10}')
  imprimirItens(vendas)
  print('\n')
```

# 5. Custo médio do frete por loja


```python
for i in range(len(lojas)):
  cm_frete.append(calcularMedia(lojas[i],'Frete'))
  print(f'Custo médio do frete da loja {i+1}: {cm_frete[i]}')
```

# *Visualização dos dados*

## 1. Faturamento

```python
cores = ['#69b3a2', '#437c90', '#f45b69', '#faae7b']
plt.bar(lojas_descricao, faturamentos, color=cores)
plt.ylim(1000000,1600000)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'R$ {x/1000000:.2f}M'))
plt.axhline(media, color='purple', linestyle='dashed', linewidth=2, label=f'Média ({media:.2f})')
plt.title('Faturamento das lojas')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle=':', color='gray', alpha=0.5)
plt.show()
```

## 2. Vendas por categoria

```python
# Unificando todas as categorias, para gerar vários gráficos com uma única legenda
categoria_vendas_ordenada = []
for i in range(len(categoria_vendas)):
  cv_loja = [list(x) for x in zip(*sorted(zip(categoria_vendas[i][0], categoria_vendas[i][1])))]
  categoria_vendas_ordenada.append(cv_loja)

# configurando os parâmetros do gráfico
labels = categoria_vendas_ordenada[0][0]

# Criando a figura com 4 subgráficos
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# lista de tamanhos e títulos
sizes = []
titles = []

for i in range(len(categoria_vendas_ordenada)):
  sizes.append(categoria_vendas_ordenada[i][1])
  titles.append(f'Vendas por categoria da loja {i+1}')

for i in range(4):
    axes[i].pie(sizes[i], autopct='%1.1f%%', shadow='True', startangle=90)
    axes[i].set_title(titles[i])

fig.legend(labels, loc="upper center", bbox_to_anchor=(0.5, -0.05), ncol=4, title="Categorias")

plt.tight_layout()
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

# Simulando 4 datasets com as colunas: vendas, latitude, longitude
df1 = pd.DataFrame({'x': loja.lat,'y': loja.lon})
df2 = pd.DataFrame({'x': loja2.lat,'y': loja2.lon})
df3 = pd.DataFrame({'x': loja3.lat,'y': loja3.lon})
df4 = pd.DataFrame({'x': loja4.lat,'y': loja4.lon})

# Inicializa o mapa
m = folium.Map(location=[-23.55, -46.63], zoom_start=5, width='100%', height='800px')

# Função para adicionar camadas
def add_heat_layer(df, name):
    heat_data = df[['x', 'y']].values.tolist()
    layer = folium.FeatureGroup(name=name)
    heat_layer = HeatMap(heat_data, radius=15, blur=15)
    layer.add_child(heat_layer)
    m.add_child(layer)

# Adiciona cada camada
add_heat_layer(df1, 'Loja 1')
add_heat_layer(df2, 'Loja 2')
add_heat_layer(df3, 'Loja 3')
add_heat_layer(df4, 'Loja 4')

# Controle de camadas
folium.LayerControl().add_to(m)

# Mostrar mapa
display(m)
```
